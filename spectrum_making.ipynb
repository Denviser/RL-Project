```python
import numpy as np
import matplotlib as plt
```

```python
#Getting IQ samples of WIFI in the form {channel_num:{[I samples],[Qsamples]}\
Wifi_symbols={1:(np.array([np.sinc(2*np.pi*0.7*i) for i in range(100000)]),np.array([np.sinc(2*np.pi*0.3*i) for i in range(100000)]))}
```

```python
mean_variance_each_channel={}

def find_mean_variance_from_fft(fft_mag_spectrum,fs=40*1e6):
    total = np.sum(fft_mag_spectrum[1:])
    probabilities = fft_mag_spectrum[1:] / total
    #bins = np.arange(len(fft_mag_spectrum))
    freq_array=np.arange(1,len(fft_mag_spectrum))*fs/(len(fft_mag_spectrum)-1)
    time_array=1/freq_array
    mean = np.sum(time_array * probabilities)
    variance = np.sum(((time_array - mean) ** 2) * probabilities)
    return mean, np.sqrt(variance)

def get_mean_variance_for_each_channel(Wifi_symbols,num_points_for_fft=10000):
    mean_variance={}
    for channel in Wifi_symbols:
        fft_I_symbols=abs(np.fft.fft(Wifi_symbols[channel][0][:num_points_for_fft]))
        mean_I_symbols,std_dev_I_symbols=find_mean_variance_from_fft(fft_I_symbols)

        fft_Q_symbols=abs(np.fft.fft(Wifi_symbols[channel][1][:num_points_for_fft]))
        mean_Q_symbols,std_dev_Q_symbols=find_mean_variance_from_fft(fft_Q_symbols)
        
        mean_variance[channel]={'I':[mean_I_symbols,std_dev_I_symbols],'Q':[mean_Q_symbols,std_dev_Q_symbols]}
    return mean_variance

mean_variance_each_channel=get_mean_variance_for_each_channel(Wifi_symbols,num_points_for_fft=10000)
print(mean_variance_each_channel)

```
