```python
import math
import random
```

```python
# multi armed bandit to select next channel to scan
```

```python
# each channel is one arm
N = 100 # number of arms
arms_with_reward = {}
arm_last_played_time = {}
arm_mean_freq = {}
arm_stddev = {}
arm_rounds_played = {}
for i in range(1, N+1):
    arms_with_reward[i] = 0
    arm_last_played_time[i] = 0
    arm_mean_freq[i] = 100
    arm_stddev[i] = 1
    arm_rounds_played[i] = 0
arm_mean_freq[45] = 1
```

```python
def get_reward(T, mu, t):
    reward = 1 - (T + mu*math.ceil((t-T)/mu) - t)/mu
    return (reward)
```

```python
def selected_arm(epsilon):
    k = random.random()
    if (k < epsilon) : 
        next_arm = random.randint(1, 100)
    else:
        next_arm = max(arms_with_reward, key = arms_with_reward.get)
    return(next_arm)
```

```python
epsilon = 1.0
epsilon_decay = 0.995
num_epochs = 400

for i in range(num_epochs):
    next_arm = selected_arm(epsilon)
    reward_secured = get_reward(arm_last_played_time[next_arm], arm_mean_freq[next_arm], i)
    arm_rounds_played[next_arm] += 1
    arms_with_reward[next_arm] += (reward_secured - arms_with_reward[next_arm]) / arm_rounds_played[next_arm]
    arm_last_played_time[next_arm] = i
    epsilon = epsilon * epsilon_decay
    print(next_arm, " is selected for ", 6*arm_stddev[next_arm], " seconds")
```
